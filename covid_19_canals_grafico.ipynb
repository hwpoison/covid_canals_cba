```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv("registro_covid_n.csv")
```

```python
cdf = df[['Fecha', 'C. Activos', 'Fallecidos']]
cdf.dropna()
rng = pd.date_range('1/1/2020', periods=len(cdf[['C. Activos']]), freq='d')
```

```python
plt.title("CASOS ACTIVOS A LO LARGO DEL TIEMPO")

plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1)) 
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%m/%y"))
plt.gcf().autofmt_xdate()

plt.plot(rng, cdf[['C. Activos']], color="blue", label="Casos Positivos")
plt.plot(rng, cdf[['Fallecidos']], color="black", label="Fallecidos")

plt.xticks(rotation=50)
plt.legend()
plt.grid(which='major', linestyle='--')
```
